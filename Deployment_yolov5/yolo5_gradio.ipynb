```python
import gradio as gr
import torch
from PIL import Image, ImageDraw
from yolov5 import utils
```

**Here we have to give the path of trained model**

```python
import torch

# Path to our custom model's weights file
custom_model_path = '/content/best.pt'

# Loading the custom YOLO model
model = torch.hub.load('ultralytics/yolov5', 'custom', path=custom_model_path)

# Now, 'model' holds our custom YOLO model
```

```python
def detect_objects(image):
    results = model(image)
    labels, cords = results.xyxyn[0][:, -1].numpy(), results.xyxyn[0][:, :-1].numpy()
    return labels, cords, image
```

```python
def draw_boxes(labels, cords, image):
    image = Image.fromarray(image)
    draw = ImageDraw.Draw(image)
    for cord in cords:
        x_min, y_min, x_max, y_max = cord[:-1]
        x_min, y_min, x_max, y_max = int(x_min * image.width), int(y_min * image.height), int(x_max * image.width), int(y_max * image.height)
        draw.rectangle([x_min, y_min, x_max, y_max], outline ="red", width=2)
    return image

def predict(input_image):
    labels, cords, image = detect_objects(input_image)
    output_image = draw_boxes(labels, cords, image)
    return output_image

iface = gr.Interface(fn=predict, inputs="image", outputs="image")
iface.launch(
    debug=True)
```
